```
from transformers import pipeline
print("START")

classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

def classifier_extractor(text: str, labels: list):
    raw_results = classifier(text, labels, multi_label=True)

    final_results = []
    for i, score in enumerate(raw_results['scores']):
        if score >= 0.8:
            final_results.append({'label':raw_results['labels'][i],
                                  'score':score})
    return final_results
```

## **QUESTIONING ANSWERING**

Menemukan jawaban spesifik dalam suatu teks (konteks) berdasarkan pertanyaan yang diberikan. Model ini bersifat extractive, artinya jawaban yang dihasilkan adalah potongan eksak dari teks konteks, bukan teks yang digenerasi baru

```
from transformers import pipeline

# Inisialisasi pipeline question answering
qa_pipeline = pipeline("question-answering",
                      model="distilbert-base-cased-distilled-squad")

# Contoh input: konteks dan pertanyaan
context = """
Jakarta adalah ibu kota Indonesia yang terletak di pulau Jawa.
Kota ini memiliki populasi lebih dari 10 juta jiwa dan dikenal
dengan kemacetan lalu lintasnya.
"""
question = "Di pulau manakah Jakarta berada?"

# Mendapatkan jawaban
output = qa_pipeline(question=question, context=context)

print(f"Jawaban: {output['answer']}")        # Output: pulau Jawa
print(f"Skor confidence: {output['score']:.3f}")  # Output: 0.987
print(f"Posisi awal: {output['start']}")     # Output: 48
print(f"Posisi akhir: {output['end']}")      # Output: 58
```

## **SENTENCE SIMILARITY**

Mengukur seberapa mirip dua atau lebih kalimat dalam hal makna. Kalimat diubah menjadi vektor embedding, lalu dihitung kemiripannya menggunakan metrik seperti cosine similarity (rentang 0-1)

```
from sentence_transformers import SentenceTransformer, util

# Load model (khusus untuk sentence similarity)
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Contoh kalimat yang akan dibandingkan
kalimat1 = "Kucing itu tidur di atas sofa"
kalimat2 = "Seekor kucing sedang berbaring di kursi empuk"
kalimat3 = "Mobil merah melaju dengan cepat di jalan tol"

# Encode menjadi vektor embedding
embedding1 = model.encode(kalimat1)
embedding2 = model.encode(kalimat2)
embedding3 = model.encode(kalimat3)

# Hitung cosine similarity
similarity_12 = util.cos_sim(embedding1, embedding2)
similarity_13 = util.cos_sim(embedding1, embedding3)

print(f"Kemiripan kalimat 1 & 2: {similarity_12[0][0]:.3f}")  # Output: 0.875
print(f"Kemiripan kalimat 1 & 3: {similarity_13[0][0]:.3f}")  # Output: 0.234
```

## **TEXT GENERATION**

Menghasilkan teks baru secara berkelanjutan berdasarkan prompt awal. Model autoregressive (decoder-only seperti GPT) memprediksi token berikutnya satu per satu, menggunakan output sebelumnya sebagai input berikutnya. Dikenal pula sebagai causal language modeling

```
from transformers import pipeline

# Inisialisasi pipeline text generation
generator = pipeline("text-generation", model="gpt2")

# Contoh input prompt
prompt = "Pada suatu hari, di tengah hutan yang sunyi,"

# Generate teks dengan parameter
output = generator(prompt,
                   max_length=50,          # Panjang maksimal output
                   num_return_sequences=2, # Jumlah output berbeda
                   temperature=0.8,        # Kreativitas (0-1, makin tinggi makin kreatif)
                   do_sample=True)         # Sampling untuk variasi

for i, hasil in enumerate(output):
    print(f"Generasi {i+1}: {hasil['generated_text']}")
    print("---")
```

## **TEXT RANKING**

Mengurutkan dokumen berdasarkan relevansinya terhadap suatu kueri (query). Model ini (sering disebut cross-encoder atau reranker) menerima pasangan (query, dokumen) dan menghasilkan skor relevansi. Berbeda dengan bi-encoder (sentence similarity) yang menghitung kemiripan embedding, cross-encoder lebih akurat namun lebih lambat karena memproses setiap pasangan secara independen

```
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load model cross-encoder untuk reranking
model_name = "cross-encoder/ms-marco-MiniLM-L-6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Kueri dan dokumen yang akan diranking
query = "cara membuat kue brownies coklat"
documents = [
    "Resep brownies panggang dengan coklat blok",
    "Panduan membuat kue nastar untuk lebaran",
    "Tips memilih coklat terbaik untuk brownies",
    "Cara membuat martabak manis teflon"
]

# Hitung skor relevansi untuk setiap dokumen
scores = []
for doc in documents:
    inputs = tokenizer(query, doc, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        score = model(**inputs).logits[0].item()
    scores.append(score)

# Urutkan dokumen berdasarkan skor (tertinggi ke terendah)
ranked = sorted(zip(documents, scores), key=lambda x: x[1], reverse=True)

print("Hasil ranking (tertinggi ke terendah):")
for i, (doc, score) in enumerate(ranked, 1):
    print(f"{i}. Skor {score:.3f} -> {doc[:50]}...")
```

## **TOKEN CLASSIFICATION**

Memberikan label pada setiap token (kata atau sub-kata) dalam suatu kalimat. Tugas paling umum adalah Named Entity Recognition (NER) yang mengidentifikasi entitas seperti nama orang (PER), lokasi (LOC), organisasi (ORG), dan tanggal/waktu

```
from transformers import pipeline

# Inisialisasi pipeline token classification (NER)
ner_pipeline = pipeline("token-classification",
                       model="dslim/bert-base-NER",
                       aggregation_strategy="simple")  # Menggabungkan token sub-kata

# Contoh input kalimat
text = "Elon Musk, CEO dari Tesla, mengunjungi Berlin pada bulan Januari 2024."

# Ekstraksi entitas
output = ner_pipeline(text)

print("Entitas yang ditemukan:")
for entity in output:
    print(f"- {entity['word']:12} | {entity['entity_group']:6} | Score: {entity['score']:.3f}")
```

## **ZERO-SHOT CLASSIFICATION (KLASIFIKASI TANPA DATA LATIH)**

Mengklasifikasikan teks ke dalam kategori yang tidak pernah dilihat saat pelatihan. Model memanfaatkan teknik Natural Language Inference (NLI): diberikan premis (teks input) dan hipotesis ("Teks ini membahas tentang {kategori}"), model menentukan apakah hipotesis benar atau salah. Tidak memerlukan fine-tuning atau data label sama sekali

```
from transformers import pipeline

# Inisialisasi pipeline zero-shot classification
classifier = pipeline("zero-shot-classification",
                     model="facebook/bart-large-mnli")

# Contoh teks yang akan diklasifikasi
text = "Pemerintah mengumumkan kenaikan suku bunga untuk mengendalikan inflasi."

# Definisikan kategori yang diinginkan (tanpa training)
candidate_labels = ["ekonomi", "politik", "olahraga", "teknologi", "kesehatan"]

# Lakukan klasifikasi
output = classifier(text, candidate_labels, multi_label=False)

print(f"Teks: {text}")
print(f"Hasil klasifikasi:")
for label, score in zip(output['labels'], output['scores']):
    print(f"  - {label:10} : {score:.3f}")
```

```
from gliner import GLiNER

## Load model gliner_base
model = GLiNER.from_pretrained("tomaarsen/gliner_base")

# Teks dan label yang sama
text = "Tolong kerjakan nomor 5 dan 7 untuk materi fisika kuantum serta nomor 3 untuk termodinamika."
labels = ["nomor soal", "materi soal"]

# Ekstraksi entitas
entities = model.predict_entities(text, labels, threshold=0.5)

# Tampilkan hasil
for entity in entities:
    print(f"- Entitas: '{entity['text']}' -> Tipe: '{entity['label']}'")
```
